```python
filtered_columns = [
    "firstName" , "lastName" , "companyName"  , "linkedinCompanyUrl" , "linkedinJobTitle",
    "linkedinHeadline" , "companyIndustry" , "location" , "linkedinProfileUrl" , "linkedinJobDateRange"
]

df = df[df.columns.intersection(filtered_columns)]
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=["companyName"])
```

```python
df["linkedinJobTitle"] = df["linkedinJobTitle"].fillna("Unknown Job Title")
df["linkedinCompanyUrl"] = df["linkedinCompanyUrl"].fillna("")
df["linkedinJobDateRange"] = df["linkedinJobDateRange"].fillna("")
```

```python
df["firstName"] = df["firstName"].str.title()
```
